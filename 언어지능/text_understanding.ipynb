```python
import os # 파일 경로 가져오기
import re # Regular Expression(정규 표현식: 특정 규칙을 가진 문자열을 검색, 추출, 치환) 구현
import pandas as pd
import tensorflow as tf
from tensorflow.keras import utils
```

```python
data_set = tf.keras.utils.get_file(
    fname = 'imdb.tar.gz',
    origin = 'http://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz',
    extract = True,
)

data_set
```

```python
# train_df, test_df 형태로 정리
def directory_data(directory):
  data = {}
  data["review"] = []
  for file_path in os.listdir(directory):
    with open(os.path.join(directory, file_path), "r", encoding='utf-8') as file:
      data["review"].append(file.read())
  return pd.DataFrame.from_dict(data)

# 긍정 -> 1, 부정 -> 0
def data(directory):
  pos_df = directory_data(os.path.join(directory, "pos")) # 긍정적인 글
  neg_df = directory_data(os.path.join(directory, "neg")) # 부정적인 글
  pos_df["sentiment"] = 1
  neg_df["sentiment"] = 0
  return pd.concat([pos_df, neg_df])

train_df = data(os.path.join(os.path.dirname(data_set), "aclImdb", "train"))
test_df = data(os.path.join(os.path.dirname(data_set), "aclImdb", "test"))
train_df.head()
```

```python
reviews = list(train_df['review'])
print(reviews[0])
print(reviews[1])

# 공백 기준 나누기
tokenized_reviews = [r.split() for r in reviews]
print(tokenized_reviews[0])
print(tokenized_reviews[1])

# 알파벳 개수
len_review_by_words = [len(t) for t in tokenized_reviews]
print(len_review_by_words[0])
print(len_review_by_words[1])

# 단어 개수
len_review_by_alphabet = [len(s.replace(' ', '')) for s in reviews]
print(len_review_by_alphabet[0])
print(len_review_by_alphabet[1])
```
